Function for showing schema of a database

```python
def show_database_tables_schema(database):
    cur.execute("USE {}".format(database))
    cur.execute("SHOW tables")
    tables = []
    for each in cur:
        tables.append(each[0])
    for each in tables:
        print("++++++++++++++++++++++++++++++++++++++")
        cur.execute("DESCRIBE {}".format(each))
        for each in cur:
            print(each)
        print("++++++++++++++++++++++++++++++++++++++")
```

```python
show_database_tables_schema("telecom")
```
