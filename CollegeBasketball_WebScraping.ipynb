```python
def conference_to_csv(dict):
  conference_dataframes = []
  for k, v in dict.items():
      example = pd.read_html(k, encoding='ISO-8859-1')
      example_df = example[0]
      example_df.sort_values(by="School", inplace = True)
      conference_dataframes.append(example_df)
  return conference_dataframes
        
```
